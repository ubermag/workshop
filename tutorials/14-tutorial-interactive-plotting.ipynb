# Tutorial 14: Interactive plotting

In this tutorial, we demonstrate basic building blocks how an interactive plot can be built.

```python
import k3d
import discretisedfield as df
%matplotlib inline

p1 = (-5e-9, -5e-9, -2e-9)
p2 = (5e-9, 5e-9, 10e-9)
cell = (1e-9, 1e-9, 1e-9)
mesh = df.Mesh(p1=p1, p2=p2, cell=cell)

value_fun = lambda pos: (pos[0], pos[1], pos[2])

def norm_fun(pos):
    x, y, z = pos
    if x**2 + y**2 < 5e-9**2:
        return 1
    else:
        return 0
    
field = df.Field(mesh, dim=3, value=value_fun, norm=norm_fun)
```

## `matplotlib` interactive plots

```python
field.plane('z').mpl()
```

```python
field.x.plane('z').mpl()
```

```python
field.x.plane(z=0).mpl()
```

```python
@df.interact(z=field.mesh.slider('z'))
def myplot(z):
    field.x.plane(z=z).mpl()
```

```python
@df.interact(z=field.mesh.slider('z', continuous_update=False))
def myplot(z):
    field.x.plane(z=z).mpl()
```

```python
@df.interact(z=field.mesh.slider('z', continuous_update=False),
             component=field.mesh.axis_selection())
def myplot(z, component):
    getattr(field, component).plane(z=z).mpl(figsize=(6, 6))
```

```python
@df.interact(z=field.mesh.slider('z', continuous_update=False),
             component=field.mesh.axis_selection(widget='radiobuttons', description='component'))
def myplot(z, component):
    getattr(field, component).plane(z=z).mpl(figsize=(6, 6))
```

```python
@df.interact(z=field.mesh.slider('z', continuous_update=False))
def myplot(z):
    field.plane(z=z).mpl(figsize=(6, 6))
```

## `k3d` interactive plots

```python
field.plane(z=0).z.k3d_voxels(filter_field=field.norm)
```

```python
import k3d

plot1 = k3d.plot()
plot1.display()

@df.interact(z=field.mesh.slider('z', continuous_update=True))
def myplot(z):
    field.z.plane(z=z).k3d_voxels(filter_field=field.norm,
                                  interactive=True,
                                  field=field,
                                  plot=plot1)   
```

```python
plot = k3d.plot()
plot.display()

@df.interact(z=field.mesh.slider('z', continuous_update=True),
             axis=field.mesh.axis_selection())
def myplot(z, axis):
    getattr(field, axis).plane(z=z).k3d_voxels(filter_field=field.norm,
                                               interactive=True,
                                               field=field,
                                               plot=plot)
```

```python
plot = k3d.plot()
plot.display()

@df.interact(y=field.mesh.slider('y', continuous_update=True),
             axis=field.mesh.axis_selection())
def myplot(y, axis):
    field.plane(y=y).k3d_vectors(color_field=getattr(field, axis),
                                 interactive=True,
                                 field=field,
                                 plot=plot)
```
